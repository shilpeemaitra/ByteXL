### **In this exercise we are working with different methods to read an image and converting the format using Python Imaging library (PIL)**

#a. Image storage formats

```python
# importing libraries
import cv2
import matplotlib.pyplot as plt
import imageio.v2 as imageio
import PIL #python imaging library
```

```python
#cv2
im = cv2.imread("/content/butterfly")
rgb_im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_im)
plt.title("Using OpenCV")
plt.show()
```

```python
#imageio
im = imageio.imread("/content/butterfly")
plt.imshow(im)
plt.title("Using ImageIO")
plt.show()
```

```python
# import matplotlib.pyplot as plt
im = plt.imread("/content/butterfly")
plt.imshow(im)
plt.title("Using Matplotlib")
plt.show()
```

```python
im= PIL.Image.open("/content/butterfly")
plt.imshow(im)
plt.title("Using PIL")
plt.show()
```

Writing images while changing its format

```python
# Checking the original format
im = PIL.Image.open("/content/butterfly")
print(f"PIL format: {im.format}")
im
```

```python
im=imageio.imread("/content/butterfly")
imageio.imsave("butterfly.jpeg",im) #webp to jpeg
im_jpeg = PIL.Image.open("/content/butterfly.jpeg")  #checking if the file was saved
print(f"PIL format: {im_jpeg.format}")
im_jpeg
```

```python
imageio.imsave("butterfly.png",im) #webp to png
im_png= PIL.Image.open("/content/butterfly.png")  #checking if the file was saved
print(f"PIL format: {im_png.format}") #prints the new format
im_png
```
